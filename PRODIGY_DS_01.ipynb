```python
import pandas as pd
import matplotlib.pyplot as plt

#Assuming df is already loaded
df = pd.read_csv("/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_6298256.csv")

#Using Region column for creating the Bar chart
region_counts = df['Region'].value_counts()

#Plotting the Bar chart
plt.bar(region_counts.index, region_counts.values)
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Distribution of Region")
plt.xticks(rotation=55)

# Now rotate the x-axis labels for better understanding
plt.tight_layout()
plt.show()

```
